```python
import numpy as np
import matplotlib.pyplot as plt
    
def omega_isotrope(phi,zeta):
    return 1 + np.tan(phi)**2*np.tan(zeta)**2 / 2

def omega_shadow_cone(phi,zeta):
    ss = np.sin(phi)*np.sin(zeta)
    cc = np.cos(phi)*np.cos(zeta)
    psi = np.arccos(-cc/ss)
    return np.pi * (ss**2*(psi/2+np.sin(2*psi)/4) + 2*ss*cc*np.sin(psi) + cc**2*psi) \
                 / (ss*np.sin(psi)+cc*psi)**2

ft = 20
colors = ['b','c','gold','orange','r','darkred']
zetas = np.linspace(0,np.pi/2,1000)
phis_deg = [1,10,30,50,70,89]
phis = np.pi/180*np.array(phis_deg)
plt.figure()
for i,phi in enumerate(phis):
    shadow = (zetas+phi>np.pi/2)
    plt.plot(180/np.pi*zetas[~shadow],omega_isotrope(phi,zetas[~shadow]),'-',color=colors[i],label=str(phis_deg[i]))
    plt.plot(180/np.pi*zetas[shadow],omega_shadow_cone(phi,zetas[shadow]),':',color=colors[i])
plt.axhline(1.5,linestyle='--',color='k')
plt.xlabel("$\zeta$",fontsize=ft)
plt.ylabel("$\omega$",fontsize=ft)
plt.rcParams['legend.title_fontsize'] = 'large'
plt.legend(title="$\\varphi$",loc='center left', bbox_to_anchor=(1, 0.5))
```
