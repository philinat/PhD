```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')

nz = 100
zi = 2000
zm = zi
z = np.linspace(0,zi,nz)

g = 9.81 # m/s^2
rho0 = 1 # kg/m^3
theta0 = 300 # K
Delta_theta = 5 # K
L = 50e3 # m

K = 200 # m^2/s
A = - g * Delta_theta / theta0 / L / K


def u_K_BL(z,zi,zm):
    return A/6 * ((zi-z)**3 + zi**3*(z/zm-1))
    #return -A/6 * (zi-z)*(2*zi - z)*z
def u_K_top(z,zi,zm):
    return A/6 * zi**3*(z/zm-1)



def dpdx_rho_BL(z,zi):
    return -K*A*(zi-z)
def dpdx_rho_top(z,zi):
    return 0*z

ft = 15
plt.figure(figsize=(6,4.6))
#plt.plot(np.concatenate((dpdx_rho_BL(z[BL],zi),dpdx_rho_top(z[~BL],zi))),z,color='k',label='pressure')
plt.axhline(zi/2,linestyle='--',linewidth=0.5,color='k')

BL = z<zi
plt.plot(np.concatenate((u_K_BL(z[BL],zi,zm),u_K_top(z[~BL],zi,zm))),z,label='0')

for awt in [0.1,0.5,2,10]:
    F = awt /K
    B = ( -A/F/2*zi**2*np.exp(-F*zi) -A/F**2*zi*np.exp(-F*zi) +A/F**3*(1-np.exp(-F*zi))) / \
            (zm-zi + (1-np.exp(-F*zi))/F)
    def u_KF_BL(z,zi,zm):
        #return 1/F*(-A/2*(zi-z)**2 +B+ (A/2*zi**2 - B)*np.exp(-F*z)) \
        #            -A/F**2*(zi-z-zi*np.exp(-F*z)) \
        #            -A/F**3*(1-np.exp(-F*z))
        return A*((zi/2+1/F)/F*(z-zi*(1-np.exp(-F*z))/(1-np.exp(-F*zi))) + z/F*(zi-z)/2)
    def u_KF_top(z,zi,zm):
        return B*(z-zm)
    
    def u_F_BL(z,zi,zm):
        return -A/F *(zi-z)**2/2
    def u_F_top(z,zi,zm):
        return 0*z
    plt.plot(np.concatenate((u_KF_BL(z[BL],zi,zm),u_KF_top(z[~BL],zi,zm))),z,label=str(awt))
    #plt.plot(np.concatenate((u_F_BL(z[BL],zi,zm),u_F_top(z[~BL],zi,zm))),z)#,label=str(awt))

plt.xlabel("$u_e$ (m/s)",fontsize=ft)
plt.ylabel("$z$ (m)",fontsize=ft)

plt.rcParams['legend.title_fontsize'] = 'large'
title = "$z_i \ = "+str(zi)+"$ m  \n"
title+= "$\\Delta\\theta = "+str(Delta_theta)+"$ K  \n"
title+= "L  = "+str(int(L/1000))+" km  \n"
title+= "$\\theta_0 = "+str(theta0)+"$ K  \n"
title+= "$\\rho_0 = "+str(rho0)+"$ kg/m³ \n"
title+= " ↓  \n"
title+= "$\\Delta p (0) = "+str(rho0* K*A*zi *L /100)+"$ hPa\n"
title+= "\n"
title+= "$K_u = "+str(K)+"$ m²/s \n"
title+= "$\\alpha_t w_t$ (m/s)"
plt.legend(title=title,loc='center left',alignment='left' ,bbox_to_anchor=(1, 0.5))
```
