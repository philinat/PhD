# Solutions à l'équilibre d'une brise de pente d'épaisseur constante



## Equations d'évolution verticale de $w_s$ et $\theta_s$:
$$\begin{aligned}
{\begin{aligned}
{w_{s}\partial_{z}w_{s}}&={\,B_{s}-\,w_{s}^{2}\left(\epsilon_{s}+\frac{C_{d}}{D\,\sin\left(\varphi\right)}\right)}\\
{\partial_{z}\theta_{s}}&={\epsilon_{s}\left(\theta_{e}-\theta_{s}\right)\,+\,\frac{F_{H}^{\perp}}{D\,w_{s}}}	
\end{aligned}}&{\,\,\,where\,\,\,\,\,\,\begin{aligned}
{B_{s}}&={g\frac{\theta_{s}-\theta_{e}}{\theta_{e}}\sin^{2}\left(\varphi\right)}\\
{F_{H}^{\perp}}&={\frac{H\cos\left(\varphi\right)}{\rho \,c_{p}}}	
\end{aligned}\,}	
\end{aligned}$$

- $\varphi$ = slope
- $B_{s}$ = effective Boyancy of the Slope wind
- $F_{H}^{\perp}$ = Heat Flux normal to the slope
- $D$ = Depth of the surface layer
- $C_d$ = Drag Coeficient on a flat surface

## Brise à l’équilibre :

Soit $\Delta = \theta_{s}-\theta_{e}$, une brise à l'équilibre est telle que :
$$\begin{aligned}
\partial_{z}w_{s}&=0\\
\partial_{z}\Delta &=0  \iff \partial_{z}\theta_{s}=\partial_{z}\theta_{e} \\
\end{aligned}$$
Nous allons donc chercher à déterminer $w_s$ et $\Delta$ en fonction des autres paramètres, et les équations d'équilibre sont :

$$\begin{aligned}
{\begin{aligned}
\beta\Delta &={\,\,w_{s}^{2}\left(\epsilon_{s}+d\right)}\\
{\partial_{z}\theta_{s}}&={-\epsilon_{s}\Delta\,+\,\frac{F_{H}^{\perp}}{D\,w_{s}}}	
\end{aligned}}&{\,\,\,where\,\,\,\,\,\,\begin{aligned}
{\beta}&={\frac{g}{\theta_{e}}\sin^{2}\left(\varphi\right)}\\
{d}&={\frac{C_{d}}{D\,\sin\left(\varphi\right)}}	
\end{aligned}\,}	
\end{aligned}$$

# 3 niveaux de complexité pour $\epsilon_s$ : 

## 1) Pas d’entrainement :  $\epsilon_{s}\,=\,0$

$$\left\{
\begin{aligned}
\beta\Delta &= \,w_{s}^{2}d\\
\partial_{z}\theta_{e} &= \,\frac{F_{H}^{\perp}}{D\,w_{s}}
\end{aligned}
\right.
\longrightarrow 
\left\{
\begin{aligned}
\Delta&= \frac{w_{s}^{2}d}{\beta}\\
w_{s} &= \frac{F_H^{\perp}}{D\partial_{z}\theta_e} 
\end{aligned}
\right.$$



## 2) Entrainement par accélération : $\epsilon_{s}\,=\,\frac{B_{s}}{2w_{s}^{2}}$
$$
\left\{\begin{aligned}
\beta\Delta &= w_{s}^{2}\frac{\beta\Delta}{2 w_{s}^{2}} + \,w_{s}^{2}d\\
\partial_{z}\theta_{e} &= -\frac{\beta\Delta^2}{2 w_{s}^{2}} + \,\frac{F_{H}^{\perp}}{D\,w_{s}}
\end{aligned}\right.
\longrightarrow 
\left\{\begin{aligned}
\Delta&= \frac{2 w_{s}^{2}d}{\beta}\\
\partial_{z}\theta_{e} &= -\frac{2 w_{s}^{2}d^2}{\beta} \,\frac{F_{H}^{\perp}}{D\,w_{s}}
\end{aligned}\right.
\longrightarrow 
w_{s}^3 +  \partial_{z}\theta_{e}\frac{\beta}{2 d^2} w_{s} - \frac{F_{H}^{\perp}}{D}\frac{\beta}{2 d^2} = 0\\
$$
On résout $w_s$ par la méthode de Cardan : 
$$
\longrightarrow 
w_{s} = \sqrt[3]{\frac{-q+r}{2}} + \sqrt[3]{\frac{-q-r}{2}}
\left\{\begin{aligned}
q &= - \frac{F_{H}^{\perp}}{D}\frac{\beta}{2 d^2}\\
r &= \sqrt{q^2+\frac{4}{27}\left(\partial_{z}\theta_{e}\frac{\beta}{2 d^2} \right)^3 }\\
\end{aligned}\right.
$$



## 3) Entrainement avec détrainement spontané ( si anabatique instable ) : $\epsilon_{s}\,=\,\frac{B_{s}}{2w_{s}^{2}}+\epsilon_{sp}$

L'entrainement et le détrainement spontanné ($\epsilon_{sp}$ et $\delta_{sp}$ ) sont liés à l'émission de thermique comme en plaine par instabilité de la couche de surface :
$$\epsilon_{sp}\,=\frac{\delta_{sp}}{2}=\,\dot{\frac{\alpha}{2}}\frac{f_{t_{0}}}{f_{s}}=\dot{\frac{\alpha}{2}}\frac{\rho\,\alpha_{t_{0}}C_{w_{0}}w^{*}}{\rho\,\alpha_{s}\,w_{s}}=\frac{\alpha_{t_{0}}C_{w_{0}}w^{*}\cos\left(\varphi\right)}{2D\,w_{s}}$$

L'échelle de vitesse des thermiques $w^{*}$ s'exprime en fonction d'une hauteur de couche limite $h^{*}$, que l'on obtient dans ce modèle à l'équilibre comme la différence entre l'altitude de la brise et l'altitude où l'environement est de température égale à la brise : $\theta_s(z) = \theta_e(z+h^{*}) = \theta_e(z) + h^{*}\partial_{z}\theta_{e}$
$${w^{*}=\left(\frac{g}{\theta_{e}}\frac{H}{\rho c_{p}}h^{*}\right)^{\frac{1}{3}}\,\,\,\,\,where\,\,\,\,\,\,h^{*}=\frac{\theta_{s}-\theta _{e}}{\partial_{z}\theta_{e}}}$$

Ainsi $\epsilon_{sp}$ peut s'écrire :
$$
\epsilon_{sp} = \frac{\alpha_{t_{0}} C_{w_{0}}}{2 D w_s} 
\left(\frac{F_{H}^{\perp}}{\tan^2\left(\varphi\right)} \frac{\beta\Delta}{\partial_{z}\theta_{e}}\right)^{\frac{1}{3}} 
$$
Les équations d'équilibre donnent :
$$
\left\{\begin{aligned}
\beta\Delta &= w_{s}^{2}\left(\frac{\beta\Delta}{2 w_{s}^{2}}+\epsilon_{sp} \right)+ \,w_{s}^{2}d\\
\partial_{z}\theta_{e} &= -\left(\frac{\beta\Delta}{2 w_{s}^{2}} + \epsilon_{sp}\right)\Delta + \,\frac{F_{H}^{\perp}}{D\,w_{s}}
\end{aligned}\right.
\longrightarrow 
\left\{\begin{aligned}
\beta\Delta &= 2 w_{s}^{2} \left( \epsilon_{sp} + d \right)\\
\partial_{z}\theta_{e} &= -\left( 2\epsilon_{sp} + d \right)\Delta + \,\frac{F_{H}^{\perp}}{D\,w_{s}}
\end{aligned}\right.
\longrightarrow 
\left\{\begin{aligned}
\beta\Delta &=  w_s \frac{\alpha_{t_{0}} C_{w_{0}}}{D} 
\left(\frac{F_{H}^{\perp}}{\tan^2\left(\varphi\right)} \frac{\beta\Delta}{\partial_{z}\theta_{e}}\right)^{\frac{1}{3}}  + 2 d w_{s}^{2} \\
w_{s}(\Delta) &= \frac{F_{H}^{\perp} - \Delta \, \alpha_{t_{0}} C_{w_{0}} \left(\frac{F_{H}^{\perp}}{\tan^2\left(\varphi\right)} \frac{\beta\Delta}{\partial_{z}\theta_{e}}\right)^{\frac{1}{3}}} {D (\partial_{z}\theta_{e} + d \Delta ) }
\end{aligned}\right.
$$
Puis on résout $\Delta$ en cherchant avec la méthode de Newton le zéro de la fonction :
$$f(\Delta) = \beta\Delta -  w_s(\Delta) \frac{\alpha_{t_{0}} C_{w_{0}}}{D} 
\left(\frac{F_{H}^{\perp}}{\tan^2\left(\varphi\right)} \frac{\beta\Delta}{\partial_{z}\theta_{e}}\right)^{\frac{1}{3}}  - 2 d w_{s}^{2} (\Delta)$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D
import warnings
warnings.filterwarnings('ignore')

nphi = 100
phi = np.linspace(0,np.pi/2,nphi)

g = 9.81 # m/s^2
rho0 = 1 # kg/m^3
theta0 = 300 # K
D = 20 # m
Cp = 1005 # J/kg/K
Cd = 0.005
Cw0 = 0.2
At0 = 0.15 # EDKF At0*Cw0 = 0.065

def prandtl_Ws_dTHs(grad,H,phi,isHnorm=False):
    F = H/rho0/Cp*np.ones_like(phi) if isHnorm else H/rho0/Cp*np.cos(phi)
    Ws = F / grad /D
    dTHs = theta0/g * Cd*Ws**2/D/np.sin(phi)**3
    return Ws,dTHs

def cubic_root(x):
    return np.copysign(np.abs(x) ** (1. / 3), x)

def epsifrot_Ws_dTHs(grad,H,phi,isHnorm=False):
    F = H/rho0/Cp if isHnorm else H/rho0/Cp*np.cos(phi)
    coef = np.sin(phi)**4 * D*g/2/theta0/Cd**2
    p = grad * D*coef
    q = - F * coef
    det = np.sqrt( q**2 + 4/27*p**3 ) /2
    Ws = cubic_root(-q/2+det) + cubic_root(-q/2-det)
    #dTHs = Hflux_normal*np.sin(phi) /Cd/Ws - grad*D*np.sin(phi)/Cd
    dTHs = 2*theta0/g * Cd*Ws**2/D/np.sin(phi)**3
    return Ws,dTHs

def sponta_Ws_dTHs(grad,H,phi,isHnorm=False):
    # Initialisation des variables récurrentes
    F = H/rho0/Cp if isHnorm else H/rho0/Cp*np.cos(phi)
    Cd_sin = Cd/np.sin(phi)
    F_t2_grad  = F/np.tan(phi)**2/grad
    boyancy = g/theta0*np.sin(phi)**2
    ## méthode de newton
    Ws,dTHs = np.ones_like(phi),np.ones_like(phi)
    for i in range(20):
        Cft0_wstar_cos = At0*Cw0*(F_t2_grad*boyancy*dTHs)**(1./3)
        numerateur = F-dTHs*Cft0_wstar_cos
        denominateur = D*grad+dTHs*Cd_sin
        Ws = numerateur/denominateur
        dWs = -( 4/3*Cft0_wstar_cos*denominateur + numerateur*Cd_sin)/denominateur**2
        f = boyancy * dTHs - Ws**2*Cd_sin*2/D - Ws/D*Cft0_wstar_cos
        df = boyancy - 4*Ws*dWs*Cd_sin/D - (dWs+Ws/3/dTHs)/D*Cft0_wstar_cos
        dTHs -= f/df
        if np.nanmax(np.abs(f/df))<1e-10:
            #print("Itérations Newton = ",i+1)
            break
    Ws = (F-dTHs*Cft0_wstar_cos)/(D*grad+dTHs*Cd_sin)
    return Ws,dTHs

def plot_Ws_dTHs(colors,list_H,list_grad,show_sponta = True,show_prandtl = False,show_epsifrot = False,isHnorm=False,xlim=[0,90],ylim=[-1,11]):
    ft = 15
    fig,axs = plt.subplots(ncols=2,figsize=(15,4.6))
    axs[0].axhline(0,linestyle='--',linewidth=0.5,color='k')
    axs[1].axhline(0,linestyle='--',linewidth=0.5,color='k')
    
    for iH,H in enumerate(list_H):
        for ig,grad in enumerate(list_grad):
            if show_prandtl:
                Ws,dTHs = prandtl_Ws_dTHs(grad,H,phi,isHnorm=isHnorm)
                axs[0].plot(phi*180/np.pi,Ws,color=colors[iH][ig],linestyle=':')
                axs[1].plot(phi*180/np.pi,dTHs,color=colors[iH][ig],linestyle=':')#,label=str(H)+' '+str(grad))
            if show_epsifrot:
                Ws,dTHs = epsifrot_Ws_dTHs(grad,H,phi,isHnorm=isHnorm)
                axs[0].plot(phi*180/np.pi,Ws,color=colors[iH][ig],linestyle='-.')
                axs[1].plot(phi*180/np.pi,dTHs,color=colors[iH][ig],linestyle='-.')#,label=str(H)+' '+str(grad))
            if show_sponta:
                Ws,dTHs = sponta_Ws_dTHs(grad,H,phi,isHnorm=isHnorm)
                axs[0].plot(phi*180/np.pi,Ws,color=colors[iH][ig])
                axs[1].plot(phi*180/np.pi,dTHs,color=colors[iH][ig],label=str(H)+' |  '+str(grad))
            
    axs[0].set_xlim(xlim)
    axs[0].set_ylim(ylim)
    axs[0].set_xlabel("$\\varphi$ (°)",fontsize=ft)
    axs[0].set_ylabel("$w_s$ (m/s)",fontsize=ft)

    axs[1].set_xlim(xlim)
    axs[1].set_ylim(ylim)
    axs[1].set_xlabel("$\\varphi$ (°)",fontsize=ft)
    axs[1].set_ylabel("$\\Delta $ (K)",fontsize=ft)

    plt.rcParams['legend.title_fontsize'] = 'large'
    title = "$D \ = "+str(D)+"$ m  \n"
    title+= "$C_d = "+str(Cd)+"$  \n"
    title+= "$\\theta_0 = "+str(theta0)+"$ K  \n"
    title+= "$\\rho_0 = "+str(rho0)+"$ kg/m³ \n"
    title+= "$At_0 = "+str(At0)+"$  \n"
    title+= "$Cw_0 = "+str(Cw0)+"$  \n"
    #title+= " ↓  \n"
    title+= "\n"
    title+= ("$H \ \cos(\\varphi)$" if isHnorm else "    $H$    ")+" |   $\partial_z\\theta_e$ \n"
    title+= "   W/m²       K/m "
    plt.legend(title=title,loc='center left',alignment='left' ,bbox_to_anchor=(1, 0.5),fancybox=True,shadow=True)
    
    
    
    legend_elements = [Line2D([0], [0], color='w',linestyle='-', lw=1, label="$\\epsilon_s $ =  ")]#Model for 
    if show_prandtl: legend_elements.append( Line2D([0], [0], color='k',linestyle=':', lw=2, label='$0$') )
    if show_epsifrot:legend_elements.append( Line2D([0], [0], color='k',linestyle='-.',lw=2, label='$\\frac{B_s}{2w_s^{2}}$') )
    if show_sponta:  legend_elements.append( Line2D([0], [0], color='k',linestyle='-', lw=2, label='$\\frac{B_s}{2w_s^{2}}+\epsilon_{sp}$'))
    fig.legend(handles=legend_elements,loc='upper center',bbox_to_anchor=(0.5, 1.05),fancybox=True,shadow=True, ncol=1+len(legend_elements),fontsize=1.*ft)#,title_fontsize='xx-large')
```

## Comparaison des 3 solutions pour différentes stabilitées 

```python
colors = [['darkviolet','b','g','orange','r']]
list_H,list_grad = [200],[0.02,0.01,0.004,0.002,0.001]
plot_Ws_dTHs(colors,list_H,list_grad,show_prandtl = True,show_epsifrot = True,isHnorm=True)
```

```python
colors = [['darkviolet'],['b'],['g'],['orange'],['r']]
list_H,list_grad = [20,40,100,200,400],[0.002]
plot_Ws_dTHs(colors,list_H,list_grad,show_prandtl = True,show_epsifrot = True,isHnorm=True)
```
