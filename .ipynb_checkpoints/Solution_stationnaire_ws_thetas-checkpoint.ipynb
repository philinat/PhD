```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')

nphi = 100
phi = np.linspace(0,np.pi/2,nphi)

g = 9.81 # m/s^2
rho0 = 1 # kg/m^3
theta0 = 300 # K
D = 20 # m
Cp = 1005 # J/kg/K
Cd = 0.005
Cw0 = 0.2
At0 = 0.15 # EDKF At0*Cw0 = 0.065

def cubic_root(x):
    return np.copysign(np.abs(x) ** (1. / 3), x)

def epsifrot_Ws_dTHs(grad,H,phi):
    Hflux_normal = H/rho0/Cp*np.cos(phi)
    coef = np.sin(phi)**4 * D*g/2/theta0/Cd**2
    p = grad * D*coef
    q = - Hflux_normal * coef
    det = np.sqrt( q**2 + 4/27*p**3 ) /2
    Ws = cubic_root(-q/2+det) + cubic_root(-q/2-det)
    #dTHs = Hflux_normal*np.sin(phi) /Cd/Ws - grad*D*np.sin(phi)/Cd
    dTHs = 2*theta0/g * Cd*Ws**2/D/np.sin(phi)**3
    return Ws,dTHs

def prandtl_Ws_dTHs(grad,H,phi):
    Hflux_normal = H/rho0/Cp*np.cos(phi)
    Ws = Hflux_normal / grad /D
    dTHs = theta0/g * Cd*Ws**2/D/np.sin(phi)**3
    return Ws,dTHs

def sponta_Ws_dTHs(grad,H,phi):
    # Initialisation des variables récurrentes
    F = H/rho0/Cp#*np.cos(phi)
    Cd_sin = Cd/np.sin(phi)
    F_t2_grad  = F/np.tan(phi)**2/grad
    boyancy = g/theta0*np.sin(phi)**2
    # initialisation de la méthode du point fixe
    #Ws,dTHs = epsifrot_Ws_dTHs(grad,H,phi)
    Ws,dTHs = np.ones_like(phi),np.ones_like(phi)
    #Ws,dTHs = np.zeros_like(phi),np.zeros_like(phi)
    Bs = boyancy * dTHs
    for i in range(0):
        Cft0_wstar_cos = At0*Cw0*(F_t2_grad*Bs)**(1./3)
        Ws = (F-dTHs*Cft0_wstar_cos)/(D*grad+dTHs*Cd_sin)
        #Ws = F*np.exp(-dTHs*Cft0_wstar_cos/F/nF)/(D*grad+dTHs*Cd_sin)
        #Ws = F*(1+np.tanh(-dTHs*Cft0_wstar_cos/F))/(D*grad+dTHs*Cd_sin)
        Bs = Ws**2*Cd_sin*2/D + Ws/D*Cft0_wstar_cos
        
        #Bs,Ws = Ws**2*Cd_sin*2/D + Ws/D*Cft0_wstar_cos , F*(1+np.tanh(-dTHs*Cft0_wstar_cos/F))/(D*grad+dTHs*Cd_sin)
        dTHs = Bs/boyancy
        
    ## méthode de newton
    for i in range(20):
        Cft0_wstar_cos = At0*Cw0*(F_t2_grad*boyancy*dTHs)**(1./3)
        numerateur = F-dTHs*Cft0_wstar_cos
        denominateur = D*grad+dTHs*Cd_sin
        Ws = numerateur/denominateur
        dWs = -( 4/3*Cft0_wstar_cos*denominateur + numerateur*Cd_sin)/denominateur**2
        f = boyancy * dTHs - Ws**2*Cd_sin*2/D - Ws/D*Cft0_wstar_cos
        df = boyancy - 4*Ws*dWs*Cd_sin/D - (dWs+Ws/3/dTHs)/D*Cft0_wstar_cos
        dTHs -= f/df
    Ws = (F-dTHs*Cft0_wstar_cos)/(D*grad+dTHs*Cd_sin)
    return Ws,dTHs
        
ft = 15
#plt.figure(figsize=(6,4.6))
fig,axs = plt.subplots(ncols=2,figsize=(15,4.6))
#plt.plot(np.concatenate((dpdx_rho_BL(z[BL],zi),dpdx_rho_top(z[~BL],zi))),z,color='k',label='pressure')
plt.axhline(0,linestyle='--',linewidth=0.5,color='k')

colors = [['b','g'],['orange','r']]

show_prandtl = False
show_epsifrot = False
for iH,H in enumerate([100,500]):
    for ig,grad in enumerate([0.01,0.001]):
        Ws,dTHs = sponta_Ws_dTHs(grad,H,phi)
        axs[0].plot(phi*180/np.pi,Ws,color=colors[iH][ig])
        axs[1].plot(phi*180/np.pi,dTHs,color=colors[iH][ig],label=str(H)+' |  '+str(grad))
        if show_prandtl:
            Ws,dTHs = prandtl_Ws_dTHs(grad,H,phi)
            axs[0].plot(phi*180/np.pi,Ws,color=colors[iH][ig],linestyle='--')
            axs[1].plot(phi*180/np.pi,dTHs,color=colors[iH][ig],linestyle='--')#,label=str(H)+' '+str(grad))
        if show_epsifrot:
            Ws,dTHs = equi_Ws_dTHs(grad,H,phi)
            axs[0].plot(phi*180/np.pi,Ws,color=colors[iH][ig],linestyle='-.')
            axs[1].plot(phi*180/np.pi,dTHs,color=colors[iH][ig],linestyle='-.')#,label=str(H)+' '+str(grad))
        
axs[0].set_xlim([0,90])
axs[0].set_ylim([0,10])
axs[0].set_xlabel("$\\varphi$ (°)",fontsize=ft)
axs[0].set_ylabel("$w_s$ (m/s)",fontsize=ft)

axs[1].set_xlim([0,90])
axs[1].set_ylim([0,10])
axs[1].set_xlabel("$\\varphi$ (°)",fontsize=ft)
axs[1].set_ylabel("$\\theta'_s $ (K)",fontsize=ft)

plt.rcParams['legend.title_fontsize'] = 'large'
title = "$D \ = "+str(D)+"$ m  \n"
title+= "$C_d = "+str(Cd)+"$  \n"
title+= "$\\theta_0 = "+str(theta0)+"$ K  \n"
title+= "$\\rho_0 = "+str(rho0)+"$ kg/m³ \n"
title+= "$At_0 = "+str(At0)+"$  \n"
title+= "$Cw_0 = "+str(Cw0)+"$  \n"
title+= " ↓  \n"
title+= " H       |     $\partial_z\\theta_e$ \n"
title+= "W/m²       K/m \n"
plt.legend(title=title,loc='center left',alignment='left' ,bbox_to_anchor=(1, 0.5))
```
